### Mtodos de grayscale

* **Lightness**: mdia das cores menos proeminentes com as mais proeminentes:
```python
(max(r, g, b) + min(r, g, b)) / 2
```

* **Average**: mdia aritmtica simples:
```python
(r + g + b) / 3
```

* **Luminosity**: mdia ponderada onde cada peso  dado de acordo com a percepo humana:

```python
0.2126 * r + 0.7152 * g + 0.0722 * b
```



## Grayscale em tons de cinza



#Metodo Avarage (iterativo):

````python

(l, c, p) = img_np.shape

img_avg = np.zeros(shape=(l, c), dtype=np.uint8)
for i in range(l):
    for j in range(c):
        r = float(img_np[i, j, 0])
        g = float(img_np[i, j, 1])
        b = float(img_np[i, j, 2])
        
        img_avg[i, j] = (r + g + b) / 3
        

plt.imshow(img_avg, cmap = 'gray')
print(img_avg.shape)
````

Metodo Average (numpy com np.average)


````python
img_av2 = np.average(img_np[:, :, :3], axis=2)
plt.imshow(img_av2, cmap = 'gray')
````

Mtodo **Average** (NumPy, com slices):

````python
img_np_float64 = img_np[:, :, :3].astype(np.float64)
# Soma cada pixel das pginas 0, 1 e 2 e divide por 3.
# Obs.: Lembrar de converter para np.uint8!
img_avg3 = ((img_np_float64[:, :, 0] + img_np_float64[:, :, 1] + img_np_float64[:, :, 2]) / 3).astype(np.uint8)
plt.imshow(img_avg3, cmap='gray')
````
